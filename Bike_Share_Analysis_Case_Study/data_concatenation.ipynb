# Script To Load & Concatenate All 2021 DataFrames Together

```python
df = pd.read_csv(r'D:\Datasets\Divvy_Trips_Data\Divvy_Trips_2021\202101-divvy-tripdata.csv')

# we use a combo of f-strings and r-strings
for month in range(2,13):
    if month < 10:
        f = fr'D:\Datasets\Divvy_Trips_Data\Divvy_Trips_2021\20210{month}-divvy-tripdata.csv'
    else:
        f = fr'D:\Datasets\Divvy_Trips_Data\Divvy_Trips_2021\2021{month}-divvy-tripdata.csv'
    df2 = pd.read_csv(f)
    df = pd.concat([df, df2], ignore_index=True)
df
```

```python
df.to_csv(r'D:\Datasets\Divvy_Trips_Data\Divvy_Trips_2021\2021_all_data.csv')
```
